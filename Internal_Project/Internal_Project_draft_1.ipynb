```python
# Clean column names
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('[^A-Za-z0-9_]', '')
df.columns
```

```python
# Define age transformation function
def age_to_midpoint(age_category):
    if pd.isnull(age_category):
        return np.nan
    elif "or older" in age_category:
        return 80  # Upper bound for the last category
    else:
        lower, upper = map(int, age_category.replace("Age", "").replace("or older", "").split("to"))
        return (lower + upper) / 2

# Apply age transformation
df['Age_Midpoint'] = df['AgeCategory'].apply(age_to_midpoint)

# Impute missing values for numerical columns with median
numerical_cols = df.select_dtypes(include=['float64']).columns
df[numerical_cols] = df[numerical_cols].apply(lambda col: col.fillna(col.median()), axis=0)

# Impute missing values for categorical columns with mode
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].apply(lambda col: col.fillna(col.mode()[0]), axis=0)

# Drop columns with excessive missing values (optional, if a threshold is needed)
# threshold = 0.5  # Drop columns with more than 50% missing values
# heart_data = heart_data.loc[:, heart_data.isnull().mean() < threshold]

# Verify cleaning
cleaned_summary = df.isnull().sum()

# Save the cleaned dataset (optional)
output_path = '/mnt/data/Cleaned_Heart.csv'
df.to_csv('cleaned_data_heart.csv', index=False)
```

```python
# Scatterplot: Age vs. BMI
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age_Midpoint', y='BMI', hue='Sex', alpha=0.6)
plt.title('Scatterplot of Age vs. BMI by Sex')
plt.xlabel('Age (Midpoint)')
plt.ylabel('BMI')
plt.legend(title='Sex')
plt.show()
```

```python
# Bar chart: General Health distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='GeneralHealth', hue='Sex', palette='pastel')
plt.title('General Health Distribution by Sex')
plt.xlabel('General Health')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Sex')
plt.show()
```

```python
# Create a grouped data frame for stacking
activity_data = df.groupby(['Age_Midpoint', 'PhysicalActivities']).size().reset_index(name='Count')

# Pivot the data to create a stacked format
activity_pivot = activity_data.pivot(index='Age_Midpoint', columns='PhysicalActivities', values='Count').fillna(0)

# Stacked bar chart
activity_pivot.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')
plt.title('Physical Activities by Age')
plt.xlabel('Age (Midpoint)')
plt.ylabel('Count')
plt.legend(title='Physical Activity')
plt.show()
```

```python
# Histogram: Sleep Hours
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='SleepHours', kde=True, bins=15, color='blue', alpha=0.7)
plt.title('Distribution of Sleep Hours')
plt.xlabel('Sleep Hours')
plt.ylabel('Count')
plt.show()
```

```python
# Group data for averages by age
health_days = df.groupby('Age_Midpoint')[['PhysicalHealthDays', 'MentalHealthDays']].mean().reset_index()

# Dual-axis line chart
fig, ax1 = plt.subplots(figsize=(12, 6))
ax2 = ax1.twinx()

ax1.plot(health_days['Age_Midpoint'], health_days['PhysicalHealthDays'], label='Physical Health Days', color='blue', marker='o')
ax2.plot(health_days['Age_Midpoint'], health_days['MentalHealthDays'], label='Mental Health Days', color='orange', marker='o')

ax1.set_xlabel('Age (Midpoint)')
ax1.set_ylabel('Average Physical Health Days', color='blue')
ax2.set_ylabel('Average Mental Health Days', color='orange')
plt.title('Average Physical and Mental Health Days by Age')
fig.tight_layout()
plt.show()
```

```python
# Pie chart: Smoker Status
smoker_counts = df['SmokerStatus'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', colors=['lightblue', 'pink', 'yellow', 'purple'], startangle=140)
plt.title('Smoker Status Distribution')
plt.show()
```

```python
# Prepare pivot table
heatmap_data = df.pivot_table(index='GeneralHealth', values='PhysicalHealthDays', aggfunc='mean')

# Heatmap: General Health vs. Physical Health Days
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt='.1f', cbar_kws={'label': 'Average Physical Health Days'})
plt.title('General Health vs. Average Physical Health Days')
plt.xlabel('General Health')
plt.ylabel('Average Physical Health Days')
plt.show()
```
